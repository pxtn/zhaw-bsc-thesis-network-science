```python
#Imports of libraries
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
liste = [
    'Pascal',
    'Ralph',
    'Nils',
    'Andi',
    'Leo',
    'Nici',
    'Piri'
]
```

```python
import random

G = nx.Graph()
counter = 0

for el in liste:
    rand = random.randrange(0, 6)
    numb = random.randrange(100, 150)
#     G.add_node(el, number=numb)
    G.add_edge(el, liste[rand])
#     G.add_edge('Nils', el)
    
# G.add_node('Michi')
# G.add_node('Yannick')
# G.add_node('Lina')
print(f'G1: {G.number_of_nodes()}')

G.remove_nodes_from(list(nx.isolates(G)))
print(f'G2: {G.number_of_nodes()}')

fig = plt.figure(1)
nx.draw(G, with_labels=True)
# nx.draw(G3, with_labels=label, node_size=nodesize, width=width, node_color=color_map_2)
plt.show()

for x, y in nx.get_node_attributes(G, 'number').items():
    counter += 1

abc = len(nx.get_node_attributes(G, 'number'))

print(G.nodes['Pascal'])

print(f'Nodes: {G.number_of_nodes()}')
print(f'Attrs: {abc}')
print(f'Counter: {counter}')
```
