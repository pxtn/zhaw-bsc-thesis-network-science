```python
G = nx.DiGraph()
counter = 0

G.add_edge('Stephen Hawking', 'Person', label='ist_eine')
G.add_edge('Stephen Hawking', 'Oxford', label='geboren_in')
G.add_edge('Stephen Hawking', 'Cambridge', label='gestorben_in')
G.add_edge('Oxford', 'Stadt', label='ist_eine')
G.add_edge('Cambridge', 'Stadt', label='ist_eine')
G.add_edge('Oxford', 'Vereinigtes Knigreich', label='teil_von')
G.add_edge('Cambridge', 'Vereinigtes Knigreich', label='teil_von')
G.add_edge('Vereinigtes Knigreich', 'Land', label='ist_ein')
    
fig = plt.figure(1, figsize=(10, 10), dpi=150)
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, width=2)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G,'label'))
```
